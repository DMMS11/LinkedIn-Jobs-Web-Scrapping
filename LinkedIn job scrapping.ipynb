```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from time import sleep, strftime
from random import randint
import pandas as pd

browser = webdriver.Firefox()

url = 'https://www.linkedin.com/jobs/search?keywords=Data%20Science&location=gurg&geoId=&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0'
browser.get(url)
sleep(2)

no_of_jobs = browser.find_element_by_css_selector('h1>span').get_attribute('innerText')
no_of_jobs = int(no_of_jobs.replace('+','').replace(',',''))

'''
to show the total number of jobs 
'''

# i = 2
# while i <= int(no_of_jobs/25)+1: 
#     browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
#     i = i + 1
#     try:
#         browser.find_element_by_xpath('/html/body/div[1]/div/main/section[2]/button').click()
#         sleep(5)
#     except:
#         pass
#         sleep(5)

job_lists = browser.find_element_by_class_name('jobs-search__results-list')
jobs = job_lists.find_elements_by_tag_name('li')

#load jobs into Dataframe

job_id= []
job_title = []
company_name = []
location = []
date = []
job_link = []
JD = []
seniority = []
emp_type = []
job_func = []
industries = []

for job in jobs:
    job_id0 = job.find_element_by_css_selector('div').get_attribute('data-tracking-id')
    if len(job_id0)==0:
        job_id.append('None')
    job_id.append(job_id0)
 
    job_title0 = job.find_element_by_css_selector('h3').get_attribute('innerText')
    if len(job_title0)==0:
        job_title.append('None')
    job_title.append(job_title0)

    company_name0 = job.find_element_by_css_selector('h4').get_attribute('innerText')
    if len(company_name0)==0:
        company_name.append('None')
    company_name.append(company_name0)

    location0 = job.find_element_by_css_selector('[class="job-search-card__location"]').get_attribute('innerText')
    if len(location0)==0:
        location.append('None')
    location.append(location0)

    date0 = job.find_element_by_css_selector('div>div>time').get_attribute('datetime')
    if len(date0)==0:
        date.append('None')
    date.append(date0)

    job_link0 = job.find_element_by_css_selector('a').get_attribute('href')
    if len(job_link0)==0:
        job_link.append('None')
    job_link.append(job_link0)
    
    
for item in range(len(jobs)):

    job_func0=[]
    industries0=[]
    #/html/body/div[1]/div/main/section[2]/ul/li[1]/div/a
    job_click_path =f'/html/body/div[1]/div/main/section[2]/ul/li[{item+1}]/div/a'
    job_click = browser.find_element_by_xpath(job_click_path).click()
    sleep(5)

    JD_path = '/html/body/div[1]/div/section/div[2]/section[2]/div'
    JD0 = browser.find_element_by_xpath(JD_path).get_attribute('innerText')
    if len(JD0)==0:
        JD.append('None')
    JD.append(JD0)

    seniority_path = '/html/body/div[1]/div/section/div[2]/section[2]/ul/li[1]/span'
    seniority0 = browser.find_element_by_xpath(seniority_path).get_attribute('innerText')
    if len(seniority0)==0:
        seniority.append('None')
    seniority.append(seniority0)

    emp_type_path = '/html/body/div[1]/div/section/div[2]/section[2]/ul/li/span'
    emp_type0 = browser.find_element_by_xpath(emp_type_path).get_attribute('innerText')
    if len(emp_type0)==0:
        emp_type.append('None')
    emp_type.append(emp_type0)

    job_func_path = '/html/body/div[1]/div/section/div[2]/section[2]/ul/li[3]/span'
    job_func_elements = browser.find_elements_by_xpath(job_func_path)
    if len(job_func_elements)==0:
        job_func.append('None')
    for element in job_func_elements:
        job_func0.append(element.get_attribute('innerText'))
        job_func_final = ','.join(job_func0)
        job_func.append(job_func_final)

    industries_path = '/html/body/div[1]/div/section/div[2]/section[2]/ul/li[4]/span'
    industries_elements = browser.find_elements_by_xpath(industries_path)
    if len(industries_elements)==0:
        industries.append('None') 
    for element in industries_elements:
        industries0.append(element.get_attribute('innerText'))
        industries_final = ','.join(industries0)
        industries.append(industries_final)
    
```

```python
job_data = pd.DataFrame({'ID': job_id,
'Date': date,
'Company': company_name,
'Title': job_title,
'Location': location,
'Description': JD,
'Level': seniority,
'Type': emp_type,
'Function': job_func,
'Industry': industries,
'Link': job_link
})
```

```python
# cleaning description column
job_data['Description'] = job_data['Description'].str.replace('\n',' ')
job_data['Description']
```
